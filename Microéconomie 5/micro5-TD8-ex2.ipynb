### Exercice 2 : Choix optimal et équilibre général

Il y a deux états de la nature $\{1, 2\}$ qui se réalisent chacun avec probabilité $\pi_1 = \pi_2 = \frac{1}{2}$. L'agent $A$ a la fonction d'utilité $v^A(c) = \ln c$ et l'allocation initiale en biens contingents $\omega^A = (10, 2)$.


#### Question 1.

Représentons l'allocation initiale $\omega^A$ et la droite de certitude (en pointillés) dans l'espace des biens contingents. Ajoutons aussi la courbe d'indifférence de $A$ passant par $\omega^A$ (en bleu).

```r
x1 <- seq(0, 12, length=1000)

ωA <- c(10, 2)
vA <- function(x) log(x)
UA <- function(x1, x2) 1/2 * vA(x1) + 1/2 * vA(x2)
indiffA <- function(U) exp(2*U) / x1

plot(x = x1, y = x1, type = "n", asp=NA, xlab="Bien contingent 1", ylab="Bien contingent 2", xaxs="i", yaxs="i")
points(ωA[1], ωA[2], pch = 3)
    text(ωA[1]+.05*12, ωA[2]+.05*12, labels = parse(text='omega^A'), cex = 1.3, font = 3)
lines(x1, x1, lty=2)
lines(x1, indiffA(UA(ωA[1], ωA[2])), col = "dodgerblue", ylim = c(0, 12))
```

#### Question 2.

Il s'agit de trouver la demande en biens contingents de $A$, c'est-à-dire de résoudre le programme

\begin{equation}
    \max_{c^A} \; U^A(c^A) \qquad \text{s.c.} \; p \cdot c^A \leq p \cdot \omega^A.
\end{equation}

La fonction d'utilité espérée $U^A$ est équivalente à une Cobb–Douglas avec les poids $\pi_1 = \frac{1}{2}$ et $\pi_2 = \frac{1}{2}$, donc nous savons que la demande de $A$ s'écrit

\begin{equation}
    \boxed{ c_1^A(p) = \frac{p \cdot \omega^A}{2 p_1} }
    \quad \text{et} \quad
    \boxed{ c_2^A(p) = \frac{p \cdot \omega^A}{2 p_2} }.
\end{equation}

#### Questions 3 et 4.

Si $\frac{p_1}{p_2} < \frac{\pi_1}{\pi_2} = 1$ alors $c_1^A(p) > c_2^A(p)$, donc $A$ ne s'assure pas complètement.

Représentons par exemple le cas $\frac{p_1}{p_2} = \frac{1}{2}$. Alors $c_1^A(p) = 7$ et $c_2^A(p) = \frac{7}{2}$. La droite de budget a pour équation $p \cdot c^A = p \cdot \omega^A$, elle est représentée en trait plein noir.

```r
cA <- c(7, 7/2)

plot(x = x1, y = x1, type = "n", asp=NA, xlab="Bien contingent 1", ylab="Bien contingent 2", xaxs="i", yaxs="i")
points(ωA[1], ωA[2], pch = 3)
    text(ωA[1]+.05*12, ωA[2]+.05*12, labels = parse(text='omega^A'), cex = 1.3, font = 3)
lines(x1, x1, lty=2)
lines(x1, indiffA(UA(ωA[1], ωA[2])), col = "dodgerblue", ylim = c(0, 12))
points(cA[1], cA[2], pch = 3)
    text(cA[1]+.05*12, cA[2]+.05*12, labels = parse(text='c^A'), cex = 1.3, font = 3)
lines(x1, ωA[2] + ωA[1]/2 - x1/2)
```

#### Question 5.

$A$ s'assure totalement, c'est-à-dire $c_1^A(p) = c_2^A(p)$, quand $p_1 = p_2$.


#### Question 6.

On considère maintenant $B$ qui a la fonction d'utilité $v^B(c) = c$ et l'allocation initiale $\omega^B = (10, 6)$.

Puisque $v^B$ est linéaire (i.e. $B$ est neutre au risque), $B$ est indifférente entre $1$ unité de bien contingent 1 ou $\frac{\pi_1}{\pi_2} = 1$ unité de bien contingent 2. En effet, les deux lui rapportent un gain égal à $\pi_1$ en utilité espérée ($\pi_1 \times 1$ dans le premier cas, $\pi_2 \times \frac{\pi_1}{\pi_2}$ dans le second). De façon équivalente, $\mathsf{TMS}^B = \frac{\pi_1}{\pi_2}$.

Ainsi, si $\frac{p_1}{p_2} < \frac{\pi_1}{\pi_2}$ alors $B$ préfère n'acheter que du bien 1.

Si au contraire $\frac{p_1}{p_2} = \frac{\pi_1}{\pi_2}$ alors $B$ est indifférent entre acheter 1 ou 2. Dans ce cas sa courbe d'indifférence coïncide avec sa droite de budget, et tous les paniers de cette droite sont des optima de son point de vue.

On représente ci-dessous le cas $\frac{p_1}{p_2} = \frac{1}{2}$. On a alors $c^B = (22, 0)$.

```r
x1 <- seq(0, 25, length=1000)

ωB <- c(10, 6)
vB <- function(x) x
UB <- function(x1, x2) 1/2 * vB(x1) + 1/2 * vB(x2)
indiffB <- function(U) 2*U - x1
cB <- c(22, 0)

plot(x = x1, y = x1, type = "n", asp=NA, xlab="Bien contingent 1", ylab="Bien contingent 2", xaxs="i", yaxs="i")
points(ωB[1], ωB[2], pch = 3)
    text(ωB[1]+.05*25, ωB[2]+.05*25, labels = parse(text='omega^B'), cex = 1.3, font = 3)
lines(x1, x1, lty=2)
lines(x1, indiffB(UB(ωB[1], ωB[2])), col = "darkorange", ylim = c(0, 25))
points(cB[1], cB[2], pch = 3)
    text(cB[1]+.05*25, cB[2]+.05*25, labels = parse(text='c^B'), cex = 1.3, font = 3)
lines(x1, ωB[2] + ωB[1]/2 - x1/2)
```

#### Question 7.

Nous avons vu la demande de $A$ à la question 2, et discuté celle de $B$ à la question 6. Formellement pour $B$, 

\begin{equation}
    \boxed{ c_1^B(p) = \begin{cases} 
        0 &\text{si $p_1 > p_2$} \\
        \frac{p \cdot \omega^B}{p_1} &\text{si $p_1 < p_2$}
        \end{cases} }
    \quad \text{et} \quad
    \boxed{ c_2^B(p) = \begin{cases} 
        \frac{p \cdot \omega^B}{p_2} &\text{si $p_1 > p_2$} \\
        0 &\text{si $p_1 < p_2$}
        \end{cases} },
\end{equation}

et si $p_1 = p_2$ il faut simplement $c_1^B + c_2^B = \omega_1^B + \omega_2^B$.

Supposer que l'équilibre est interne implique donc de supposer $p_1 = p_2$, car autrement l'équilibre serait en coin pour répondre à la demande de $B$. 
En réinjectant le rapport des prix $\frac{p_1}{p_2} = 1$ dans les demandes de $A$, on obtient

\begin{equation}
    c_1^A = c_2^A = \frac{\omega_1^A + \omega_2^A}{2} \overset{\text{A.N.}}{=} 6.
\end{equation}

Avec les conditions de ressources on en déduit $c_1^B = 14$ et $c_2^B = 2$.

$A$ s'est donc assuré complètement en achetant 4 unités de bien 2 à $B$, contre 4 unités de bien 1. On remarque que $A$ sort strictement gagnant de cet échange ($A$ est averse au rsique et son allocation finale est moins risquée mais de même espérance que son allocation initiale) mais pas $B$ (son allocation finale a même espérance que son allocation initiale).

#### Question 8.

On suppose maintenant $v^B(c) = \ln c$. Les demandes de $B$ sont alors identiques à celles de $A$ (donc mêmes réponses à la question 6).

L'équilibre du marché du bien contingent 1 s'écrit

\begin{equation}
    \frac{p \cdot \omega^A}{2p_1} + \frac{p \cdot \omega^B}{2p_1} = \omega_1
    \iff \boxed{ \frac{p_1}{p_2} = \frac{\omega_2}{\omega_1} } \overset{\text{A.N.}}{=} \frac{2}{5}.
\end{equation}

On réinjecte ce rapport de prix dans les demandes des agents pour obtenir les allocations d'équilibre :

\begin{equation}
    \boxed{ c_1^A = \frac{15}{2}, \quad c_2^A = 3, \qquad c_1^B = \frac{25}{2}, \quad c_2^B = 5 }.
\end{equation}
