## TD6 : Risque I

### Exercice 1 : Accroissement du risque

#### Question 1.

L'agent $E$ dispose d'un niveau de richesse égal à 4000. Avec probabilité 1/2, le projet de transport réussit et il gagne 8000 de plus ; sinon, aussi avec probabilité 1/2, il ne gagne rien de plus. 

Notons que l'espérance de gain du projet est alors $\frac{1}{2} \times 8000 + \frac{1}{2} \times 0 = 4000$.

Formellement, $E$ fait donc face à la loterie $x = (12000, 4000 ; \frac{1}{2}, \frac{1}{2})$. 
Par définition, l'utilité espérée du projet est alors

\begin{aligned}
    U^E(x)
    &= \frac{1}{2} v^E(12000) + \frac{1}{2} v^E(4000) \\
    &= \frac{1}{2} \sqrt{12000} + \frac{1}{2} \sqrt{4000} = 86{,}4.
\end{aligned}


#### Question 2a.

Avec l'incertitude sur le prix de vente, l'espérance de gain du projet devient

\begin{equation}
    \frac{1}{4} \times 1{,}5 \times 8000 + \frac{1}{4} \times 0{,}5 \times 8000 + \frac{1}{2} \times 0
    = 4000.
\end{equation}

Notons qu'elle n'a pas changé par rapport à la situation où le prix est fixe.


#### Question 2b.

Représentation de la nouvelle lotterie $y = (16000, 8000, 4000 ; \frac{1}{4}, \frac{1}{4}, \frac{1}{2})$.

#### Question 2c.

L'utilité espérée associée à la nouvelle lotterie $y$ est 

\begin{equation}
    U^E(y) = \frac{1}{4} \sqrt{16000} + \frac{1}{4} \sqrt{8000} + \frac{1}{2} \sqrt{4000} = 85{,}6.
\end{equation}


#### Question 2d.

On constate que le risque supplémentaire sur les prix se traduit par une utilité inférieure pour $E$.

C'était attendu : la fonction d'utilité $v^E$ est concave, donc $E$ est averse au risque. Or $x$ et $y$ ont même espérance (puisqu'elles ont la même espérance de gain) mais $y$ est plus risquée. Donc $E$ préfère $x$ à $y$.
