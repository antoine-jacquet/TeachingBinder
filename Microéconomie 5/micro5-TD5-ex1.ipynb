# Microéconomie 5
## TD5 : Marché de droits à polluer

### Exercice 1

On considère une économie avec deux biens : le numéraire (bien 1, prix $p_1 = 1$ par hypothèse) et les droits à polluer (bien 2, en quantité totale $N$, prix $p_2 = p$). Les dotations initiales en bien 1 sont $\omega^A_1 = 6$ et $\omega^B_1 = 4$. Les agents ont les préférences

\begin{aligned}
    U^A(x^A) &= \ln(x_2^A) + x_1^A \\
    U^B(x^B) &= \ln(x_2^B) + x_1^B.
\end{aligned}

On va s'intéresser à ce qui se passe quand on distribue les droits (le bien 2) différemment aux agents $A$ et $B$. Par exemple, on peut choisir de donner initialement tous les droits à $A$, c'est-à-dire de choisir les allocations initiales $\omega^A_2 = N$ et $\omega^B_2 = 0$ ; ou à l'inverse, de tous les donner à $B$, c'est-à-dire $\omega^A_2 = 0$ et $\omega^B_2 = N$.


*Commençons par définir ces données du problème et tracer quelques courbes d'indifférence. On choisit pour illustrer $N = 10$ et on commence par le cas où $A$ possède tous les droits au départ.*

```r
N <- 10                                      # quantité totale de droits à polluer
ωA <- c(6, N)                                # dotation initiale de A
ωB <- c(4, N - ωA[2])                        # dotation initiale de B
ω1 <- ωA[1] + ωB[1]                          # dotation totale en bien 1
ω2 <- ωA[2] + ωB[2]                          # dotation totale en bien 2
UA <- function(x1, x2) log(x2) + x1          # fonction d'utilité de A
UB <- function(x1, x2) log(x2) + x1          # fonction d'utilité de B

x1 <- seq(0, ω1, length=1000)

indiffA <- function(U) exp(U - x1)
indiffB <- function(U) exp(U - x1)

niveauxA <- seq(UA(ω1/5, ω2/5), UA(4*ω1/5, 4*ω2/5), len = 4)
niveauxB <- seq(UB(ω1/5, ω2/5), UB(4*ω1/5, 4*ω2/5), len = 4)

plot(x = x1, y = x1, xlim = c(0, ω1), ylim = c(0, ω2),
     type = "n", xlab = "Bien 1", ylab = "Bien 2", xaxs = "i", yaxs = "i")
for (U in niveauxA) lines(x1, indiffA(U), col = "dodgerblue", ylim = c(0, ω2))                # courbe d'indifférence de A
for (U in niveauxB) lines(ω1 - x1, ω2 - indiffB(U), col = "darkorange", ylim = c(0, ω2))      # courbe d'indifférence de B
```

#### Question 1.

On cherche premièrement l'*équilibre concurrentiel* de cette économie quand $A$ dispose initialement de l'ensemble des droits : $\omega^A_2 = N$. Pour simplifier, supposons que l'équilibre est *interne*.

Dans ce cas, en l'absence d'externalités, l'égalité suivante est vérifiée à l'équilibre concurrentiel :

\begin{equation}
    \mathsf{TMS}^A = \mathsf{TMS}^B = \frac{p_1}{p_2} \iff x^A_2 = x^B_2 = \frac{1}{p}.
\end{equation}

L'équilibre du marché de bien 2 s'écrit donc

\begin{equation}
    \underbrace{x^A_2}_{=\frac{1}{p}} + \underbrace{x^B_2}_{=\frac{1}{p}} = \underbrace{\omega_2}_{= N} \iff p = \frac{2}{N}
\end{equation}

et fournit ainsi le prix d'équilibre des droits à polluer. (Rappellons que le bien 1 est ici le numéraire, i.e. son prix est supposé égal à 1.)

Grâce aux contraintes de budget de chaque agent, on en déduit l'équilibre en fonction des allocations initiales :

\begin{equation}
    \boxed{
    p = \frac{2}{N}
    \qquad
    x_2^A = x_2^B = \frac{N}{2}
    \qquad
    x_1^A = \omega_1^A + \frac{2}{N} \omega_2^A - 1
    \qquad
    x_1^B = \omega_1^B + \frac{2}{N} \omega_2^B - 1
    }.
\end{equation}

Notons bien que **ces formules restent valables uniquement tant que les allocations $x_1^A$ et $x_1^B$ obtenues sont positives**, c'est-à-dire tant que l'équilibre reste interne.

Si, comme on l'a supposé ici, $A$ possède initialement tous les droits (i.e. $\omega_2^A = N$) alors l'allocation d'équilibre en numéraire est

\begin{equation}
    x_1^A = \omega_1^A + 2 - 1 = 7
    \qquad
    x_1^B = \omega_1^B + 0 - 1 = 3
\end{equation}

c'est-à-dire que $B$ a échangé 1 unité de numéraire contre la moitié des droits de $A$ ($p \times \frac{N}{2} = 1$).

#### Question 2.

L'allocation d'équilibre vérifie bien $\mathsf{TMS}^A = \mathsf{TMS}^B$, et elle distribue toutes les ressources disponibles, donc elle correspond bien à un optimum de Pareto (interne).

#### Question 3.

On suppose maintenant que $B$ dispose initialement de tous les droits : $\omega_2^B = N$.

L'expression générique de l'équilibre en fonction des allocations initiales trouvée ci-dessus reste valable. Le prix d'équilibre reste donc $p = \frac{2}{N}$, les droits sont répartis en $x_2^A = x_2^B = \frac{N}{2}$, et les allocations en numéraire sont

\begin{equation}
    x_1^A = \omega_1^A + 0 - 1 = 5
    \qquad
    x_1^B = \omega_1^B + 2 - 1 = 5
\end{equation}

c'est-à-dire que $A$ a échangé 1 unité de numéraire contre la moitié des droits de $B$ (échange opposé à celui du cas précédent).

Encore une fois, cette allocation d'équilibre correspond bien à un optimum de Pareto (pour les mêmes raisons qu'à la question 2).

#### Question 4.

Remarquons que les droits sont toujours partagés de façon égale à l'équilibre, i.e. $x_2^A = x_2^B = \frac{N}{2}$. Il y a donc toujours la même quantité de pollution à l'équilibre, $\frac{N}{2}$. Ceci est une conséquence de la nature des préférences des agents : leurs fonctions d'utilité sont quasi-linéaire, ce qui entraîne une absence d'effet revenu sur leur demande en bien 2 (tant que la solution reste intérieure).

Mathématiquement, cela se traduit par le fait que les fonctions de demande brute en bien 2 des agents ne dépendent pas de l'allocation initiale $\omega$. En effet, ici $x_2^A(p) = x_2^B(p) = \frac{1}{p}$.

#### Question 5.

Finalement, que se passe-t-il si le régulateur augmente la quantité de droits à polluer $N$ ? Nous pouvons faire plusieurs commentaires.

- Le prix d'équilibre $p = \frac{2}{N}$ diminue. Ceci est cohérent avec l'idée que les droits deviennent moins rares, et donc moins chers, quand $N$ augmente.

- Les droits sont toujours partagés moitié-moitié à l'équilibre. En effet, les résultats ci-dessus ne dépendent pas de la valeur de $N$ choisie.

- Les allocations d'équilibre en bien 1 ne sont pas modifiées par une augmentation de $N$, tant que chaque agent conserve la même part relative de l'ensemble des droits (c'est-à-dire une fraction constante de $N$). En effet, le prix d'équilibre des droits diminue de telle sorte que la moitié des droits $\frac{N}{2}$ aura toujours la même valeur que 1 unité de numéraire.

*Représentons les deux équilibres discutés ci-dessus dans la boîte d'Edgeworth.*

```r
# Allocation initiale et allocation d'équilibre quand A commence avec tous les droits

p <- 2/N
ωA[2] <- N
ωB[2] <- N - ωA[2]
x1A <- ωA[1] + p*ωA[2] - 1
x2A <- N/2
x1B <- ωB[1] + p*ωB[2] - 1
x2B <- N/2

IA <- ωA
EA <- c(x1A, x2A)


# Graphe (cas où A commence avec tous les droits)

x2 <- seq(0, ω2, length=1000)
plot(x = x1, y = x1, xlim = c(0, ω1), ylim = c(0, ω2),
     type = "n", asp=NA, xlab="Bien 1", ylab="Bien 2", xaxs="i", yaxs="i")
lines(c(0, ω1), c(N/2, N/2), lwd = 2, col = "purple")                                                # optima internes
points(IA[1], IA[2], pch = 3)                                                                        # allocation initiale I_A
    text(IA[1]-.05*ω1, IA[2]-.05*ω2, labels = expression(italic("I")[italic("A")]), cex = 1.3, font = 3)
lines(x1, indiffA(UA(IA[1], IA[2])), col = "dodgerblue", ylim = c(0, ω2))                            # courbe d'indifférence de A à I_A
lines(ω1 - x1, ω2 - indiffB(UB(ω1 - IA[1], ω2 - IA[2])), col = "darkorange", ylim = c(0, ω2), lwd=4) # courbe d'indifférence de B à I_A
lines(ωA[1] + p*ωA[2] - p*x2, x2, lty = 2)                                                           # contrainte de budget à l'équilibre E_A
points(EA[1], EA[2], pch=3)                                                                          # équilibre concurrentiel E_A
    text(EA[1]+.05*ω1, EA[2]+.05*ω2, labels = expression(italic("E")[italic("A")]), cex = 1.3, font = 3)


# Allocation initiale et allocation d'équilibre quand B commence avec tous les droits

ωA[2] <- 0
ωB[2] <- N - ωA[2]
x1A <- ωA[1] + p*ωA[2] - 1
x1B <- ωB[1] + p*ωB[2] - 1

IB <- ωA
EB <- c(x1A, x2A)


# Fin du graphe (cas où B commence avec tous les droits)

points(IB[1], IB[2], pch = 3)                                                                        # allocation initiale I_B
    text(IB[1]+.05*ω1, IB[2]+.05*ω2, labels = expression(italic("I")[italic("B")]), cex = 1.3, font = 3)
lines(x1, indiffA(UA(IB[1], IB[2])), col = "dodgerblue", ylim = c(0, ω2), , lwd=4)                   # courbe d'indifférence de A à I_B
lines(ω1 - x1, ω2 - indiffB(UB(ω1 - IB[1], ω2 - IB[2])), col = "darkorange", ylim = c(0, ω2))        # courbe d'indifférence de B à I_B
lines(ωA[1] + p*ωA[2] - p*x2, x2, lty = 2)                                                           # contrainte de budget à l'équilibre E_B
points(EB[1], EB[2], pch=3)                                                                          # équilibre concurrentiel E_B
    text(EB[1]-.05*ω1, EB[2]+.05*ω2, labels = expression(italic("E")[italic("B")]), cex = 1.3, font = 3)
```

L'allocation initiale où $A$ commence avec tous les droits est $I_A$, et l'équilibre associé est $E_A$. De la même façon, l'allocation initiale où $B$ commence avec tous les droits est $I_B$, et l'équilibre associé est $E_B$.

La courbe des optima de Pareto internes est en violet. C'est un segment qui traverse la boîte d'Edgeworth, comme vu dans les TDs précédents avec des préférences quasi-linéaires. (Le segment est horizontal plutôt que vertical parce qu'ici l'utilité des agents est linéaire en bien 1 plutôt qu'en bien 2.)

Les droites en pointillés sont les droites de budget à l'équilibre.

Les courbes d'indifférence de $A$ sont en bleu, celles de $B$ en orange. On remarque que la courbe d'indifférence de $B$ passant par $I_A$ est horizontale, ainsi que celle de $A$ passant par $I_B$. C'est parce que toute allocation où $x_2^i = 0$ donne la même utilité ($-\infty$) à l'agent $i$.
