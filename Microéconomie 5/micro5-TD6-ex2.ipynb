### Exercice 2 : Diversification

Il y a 2 états de la nature, chacun se réalisant avec probabilité $0{,}5$. Les deux biens contingents, $c_1$ et $c_2$, sont les quantités obtenues respectivement dans les états 1 et 2.

On s'intéresse aux trois loteries $G_1$, $G_2$ et $G_3$ suivantes :

\begin{equation}
    G_1 = (480, 0{,}5 ; 480, 0{,}5)
    \qquad
    G_2 = (850, 0{,}5 ; 200, 0{,}5)
    \qquad
    G_3 = (1000, 0{,}5; 0, 0{,}5).
\end{equation}


#### Question 1.

Commençons par calculer leurs espérances respectives :

\begin{aligned}
    \mathbf E(G_1) &= 0{,}5 \times 480 + 0{,}5 \times 480 = \underline{480} \\
    \mathbf E(G_2) &= 0{,}5 \times 850 + 0{,}5 \times 200 = \underline{525} \\
    \mathbf E(G_3) &= 0{,}5 \times 1000 + 0{,}5 \times 0 = \underline{500}.
\end{aligned}

Représentons-les maintenant sur un graphe.

```r
G1 <- c(480, 480)
G2 <- c(850, 200)
G3 <- c(1000, 0)

c <- c(0, 1.05*max(G1, G2, G3))

plot(x = c, y = c, type = "n", xlab = "Bien 1", ylab = "Bien 2", asp=1)
abline(a=0, b=1, lty=2)                          # droite de certitude c1 = c2                
points(G1[1], G1[2], pch=19, col="red")          # loterie G1
    text(G1[1]-.05*max(c), G1[2]+.02*max(c), labels = expression("G"[1]), cex = 1.3, font = 3, col="red")
points(G2[1], G2[2], pch=19, col="blue")         # loterie G2
    text(G2[1]+.05*max(c), G2[2]+.05*max(c), labels = expression("G"[2]), cex = 1.3, font = 3, col="blue")
points(G3[1], G3[2], pch=19, col="green")        # loterie G3
    text(G3[1]-.05*max(c), G3[2]+.00*max(c), labels = expression("G"[3]), cex = 1.3, font = 3, col="green")

abline(a = G2[1] + G2[2], b = -1, col="blue")    # loteries équitables à G2
abline(a = G3[1] + G3[2], b = -1, col="green")   # loteries équitables à G3
```

En pointillés noirs j'ai représenté la droite de certitude ($c_1 = c_2$). En rouge est représentée la loterie $G_1$. Le point bleu représente la loterie $G_2$, et la droite bleue l'ensemble des loteries équitables d'espérance $\mathbb E(G_2)$. Le point vert représente enfin la loterie $G_3$, et la droite verte l'ensemble des loteries équitables d'espérance $\mathbb E(G_3)$.

En prenant comme critère qualitatif qu'une loterie est plus risquée quand elle est plus éloignée de la droite de certitude, il semble que $G_1$ soit la loterie la moins risquée, puis $G_2$, et enfin $G_3$ est la plus risquée.

#### Question 3.

Étudions maintenant les préférences de trois agents $A$, $B$ et $C$ sur ces loteries. Leurs fonctions d'utilité sur les quantités finales à consommer $c$ sont

\begin{equation}
    v^A(c) = \sqrt{c}
    \qquad
    v^B(c) = c
    \qquad
    v^C(c) = c^2
\end{equation}

donc $A$ est averse au risque, $B$ est neutre au risque , et $C$ aime le risque.
En effet, $(v^A)'' < 0$, $(v^B)'' = 0$, et $(v^C)'' > 0$.

On calcule maintenant leurs utilitées espérées respectives sur les trois loteries :

\begin{aligned}
    U^A(G_1) &= 0{,}5 \times \sqrt{480} + 0{,}5 \times \sqrt{480} \simeq 21{,}9 \\
    U^A(G_2) &= 0{,}5 \times \sqrt{850} + 0{,}5 \times \sqrt{200} \simeq 21{,}6 \\
    U^A(G_3) &= 0{,}5 \times \sqrt{1000} + 0{,}5 \times \sqrt{0} \simeq 15{,}8 \\[10pt]
    U^B(G_1) &= 0{,}5 \times 480 + 0{,}5 \times 480 = 480 \\
    U^B(G_2) &= 0{,}5 \times 850 + 0{,}5 \times 200 = 525 \\
    U^B(G_3) &= 0{,}5 \times 1000 + 0{,}5 \times 0 = 500 \\[10pt]
    U^C(G_1) &= 0{,}5 \times 480^2 + 0{,}5 \times 480^2 = 230\,400 \\
    U^C(G_2) &= 0{,}5 \times 850^2 + 0{,}5 \times 200^2 = 381\,250 \\
    U^C(G_3) &= 0{,}5 \times 1000^2 + 0{,}5 \times 0^2 = 500\,000.
\end{aligned}


#### Question 4.

On peut ainsi établir les préférences des agents sur ces loteries :

\begin{aligned}
    G_1 &\succ_A G_2 \succ_A G_3 \\[5pt]
    G_2 &\succ_B G_1 \succ_B G_3 \\[5pt]
    G_3 &\succ_C G_2 \succ_C G_1
\end{aligned}

où $x \succ_i y$ signifie que l'agent $i$ préfère strictement $x$ à $y$.

Ainsi on voit que l'ordre qualitatif établi sur les loteries à la question 2 (de la moins risquée à la plus risquée) est cohérent avec les préférences des agents : $A$ qui est averse au risque préfère les loteries moins risquées, $C$ qui aime le risque préfère les loteries plus risquées, et $B$ qui est neutre au risque préfère juste celles qui ont la meilleure espérance.

Notons aussi que l'on pouvait connaître certaines préférences sans avoir recours à l'utilité espérée des agents. Par exemple, la loterie $G_2$ a une espérance supérieure à $G_3$, *et* elle est aussi moins risquée. L'agent $A$, qui est averse au risque, préfère donc nécessairement $G_2$ à $G_3$.
