# Microéconomie 5
## TD7 : Risque II

### Exercice 1 : Attitude face au risque

Nous connaissons les préférences de deux agents $A$ et $B$ sur les deux loteries $L_1 = (w+40, 1)$ et $L_2 = (w+60, w ; \frac{1}{2}, \frac{1}{2})$. Elles sont données par

\begin{aligned}
    L_1 &\succ_A L_2 \\
    L_2 &\succ_B L_1.
\end{aligned}

Calculons l'espérance de ces lotteries : 

\begin{aligned}
    \mathbf E(L_1) &= w + 40 \\
    \mathbf E(L_2) &= w + 30.
\end{aligned}

*Notation.* Par abus de notation, pour tout $x \in \mathbf R$ on appelle aussi $x$ la loterie certaine $(x, 1)$, i.e. celle qui donne $x$ avec probabilité 1. Ainsi la *loterie $\mathbf E(L_2)$* correspond à la loterie certaine $(\mathbf E(L_2), 1)$.

Ici, la loterie $L_2$ a une espérance inférieure à celle de $L_1$, et $L_2$ est aussi plus risquée.

Puisque $B$ préfère $L_2$ à $L_1$, il faut donc nécessairement qu'il ait du goût pour le risque. Mathématiquement, on peut écrire

\begin{equation}
    U^B (L_2) > U^B(L_1) = v^B(w+40) > v^B(w+30) = U^B(\mathbf E(L_2)),
\end{equation}

c'est-à-dire que $B$ préfère la loterie risquée $L_2$ à la loterie certaine de même espérance $\mathbf E(L_2)$. Cela garantit que $B$ a du goût pour le risque.

On ne peut en revanche rien dire sur l'attitude face au risque de $A$ : il est possible qu'il ait du goût pour le risque et préfère malgré tout la lotterie moins risquée $L_1$ (parce qu'elle a une espérance plus élevée), ou bien qu'il soit neutre ou averse au risque. Mathématiquement, on a

\begin{equation}
    U^A (L_1) > U^A(L_2)
    \quad \text{et} \quad
    U^A (L_1) > U^A(\mathbf E(L_2)),
\end{equation}

mais cela ne permet pas d'établir les préférences de $A$ entre la loterie $L_2$ et la loterie certaine $\mathbf E(L_2)$.

Pour s'en convaincre, on peut considérer le cas $w = 0$ et les fonctions d'utilité $v^A_1(c) = \sqrt{c}$ et $v^A_2(c) = c^{1{,}1}$.

* Avec $v_1^A$ (concave, donc $A$ est averse au risque)
\begin{aligned}
    U^A_1(L_1) &= \sqrt{40} \simeq 6{,}3 \\
    U^A_1(L_2) &= \tfrac{1}{2} \sqrt{60} + \tfrac{1}{2} 0 \simeq 3{,}9
\end{aligned}
donc $A$ préfère $L_1$ à $L_2$.

* Avec $v_2^A$ (convexe, donc $A$ a du goût pour le risque)
\begin{aligned}
    U^A_1(L_1) &= 40^{1{,}1} \simeq 57{,}8 \\
    U^A_1(L_2) &= \tfrac{1}{2} 60^{1{,}1} + \tfrac{1}{2} 0^{1{,}1} \simeq 45{,}2
\end{aligned}
donc $A$ préfère aussi $L_1$ à $L_2$.
