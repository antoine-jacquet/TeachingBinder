### Exercice 4 : Représentations graphiques

Il y a deux états de la nature $\{1, 2\}$ qui se réalisent avec probabilité $\pi_1 = \frac{3}{4}$ et $\pi_2 = \frac{1}{4}$ respectivement. Les deux biens contingents, $c_1$ et $c_2$, sont les quantités obtenues respectivement dans les états 1 et 2.

On considère la loterie $X$ qui rapporte 50 € dans l'état 1 et 300 € dans l'état 2 : $X = (50, \frac{3}{4}; 300, \frac{1}{4})$.

On s'intéresse à trois cas pour la fonction d'utilité de l'agent $A$ :

\begin{equation}
    \text{(a)} \quad v(c)= \sqrt{c}
    \qquad
    \text{(b)} \quad v(c) = 6c
    \qquad
    \text{(c)} \quad v(c) = c^2.
\end{equation}

On souhaite simplement représenter la courbe d'indifférence de $A$ passant par $X$, la droite de certitude, et la droite des loteries équitables de même espérance que $X$, dans ces trois cas. L'équation de la courbe d'indifférence passant par $X$ s'obtient notamment en adaptant l'équation

\begin{equation}
    \pi_1 v(c_1) + \pi_2 v(c_2) = \underbrace{\pi_1 v(50) + \pi_2 v(300)}_{= U(X)} 
\end{equation}

à la fonction d'utilité $v$ considérée.

**Cas (a) : $v(c) = \sqrt{c}$.**

L'équation de la courbe d'indifférence passant par $X$ est donnée par

\begin{equation}
    \frac{3}{4} \sqrt{c_1} + \frac{1}{4} \sqrt{c_2} = \frac{3}{4} \sqrt{50} + \frac{1}{4} \sqrt{300}
    \iff
    \underbrace{\boxed{ c_2 = \left(3 \sqrt{50} + \sqrt{300} - 3 \sqrt{c_1}\right)^2 }}_\text{ pour $c_1$ tels que $3 \sqrt{50} + \sqrt{300} - 3 \sqrt{c_1} \geq 0$}.
\end{equation}

```r
X <- matrix(c(50, 3/4, 300, 1/4), 2, 2)          # matrice qui représente la loterie X
c <- seq(0, 400, length=5000)

v <- function(c) sqrt(c)
U <- 3/4*v(50) + 1/4*v(300)                      # niveau d'utilité espérée à X

plot(x = c, y = c, type = "n", xlab = "Bien 1", ylab = "Bien 2", xaxs="i", yaxs="i")
points(X[1,1], X[1,2], pch=3)                                                 # loterie X
    text(X[1, 1]+.05*max(c), X[1,2]+.05*max(c), labels = "X", cex = 1.3, font = 3)
abline(a=0, b=1, lty=2)                                                       # droite de certitude
abline(a=crossprod(X[2,], X[1,])/X[2,2], b=-X[2,1]/X[2,2])                    # loteries équitables
lines(c[3/4*v(c) < U], (4*U - 3*v(c[3/4*v(c) < U]))^2, col="red", lwd=2)      # courbe d'indifférence
```

Ici j'ai représenté :
- la droite de certitude en pointillés,
- la droite des loteries équitables passant par $X$ en noir,
- la courbe d'indifférence passant par $X$ en rouge.

On constate que dans ce cas, la courbe d'indifférence est convexe. En effet, **quand l'agent est averse au risque, sa fonction d'utilité est concave et ses courbes d'indifférence sont convexes**.

**Cas (b) : $v(c) = 6 c$.**

L'équation de la courbe d'indifférence passant par $X$ est donnée par

\begin{equation}
    \frac{3}{4} \times 6 c_1 + \frac{1}{4} \times 6 c_2 = \frac{3}{4} \times 6 \times 50 + \frac{1}{4} \times 6 \times 300
    \iff
    \boxed{ c_2 = 450 - 3 c_1 }.
\end{equation}

```r
v <- function(c) 6*c
U <- 3/4*v(50) + 1/4*v(300)                      # niveau d'utilité espérée à X

plot(x = c, y = c, type = "n", xlab = "Bien 1", ylab = "Bien 2", xaxs="i", yaxs="i")
points(X[1,1], X[1,2], pch=3)                                                 # loterie X
    text(X[1, 1]+.05*max(c), X[1,2]+.05*max(c), labels = "X", cex = 1.3, font = 3)
abline(a=0, b=1, lty=2)                                                       # droite de certitude
abline(a=crossprod(X[2,], X[1,])/X[2,2], b=-X[2,1]/X[2,2])                    # loteries équitables
lines(c, 450 - 3*c, col="red", lwd=2)                                         # courbe d'indifférence
```

Ici la courbe d'indifférence est confondue avec la droite des loteries équitables (de même espérance que $X$).

On constate que dans ce cas, la courbe d'indifférence est une droite. En effet, **quand l'agent est neutre au risque, sa fonction d'utilité est une droite et ses courbes d'indifférence sont des droites**.

**Cas (c) : $v(c) = c^2$.**

L'équation de la courbe d'indifférence passant par $X$ est donnée par

\begin{equation}
    \frac{3}{4} (c_1)^2 + \frac{1}{4} (c_2)^2 = \frac{3}{4} (50)^2 + \frac{1}{4} (300)^2
    \iff
    \underbrace{\boxed{ c_2 = \sqrt{97\,500 - 3 (c_1)^2} }}_\text{ pour $c_1$ tels que $97\,500 - 3 (c_1)^2 \geq 0$}.
\end{equation}

```r
v <- function(c) c^2
U <- 3/4*v(50) + 1/4*v(300)                      # niveau d'utilité espérée à X

plot(x = c, y = c, type = "n", xlab = "Bien 1", ylab = "Bien 2", xaxs="i", yaxs="i")
points(X[1,1], X[1,2], pch=3)                                                 # loterie X
    text(X[1, 1]+.05*max(c), X[1,2]+.05*max(c), labels = "X", cex = 1.3, font = 3)
abline(a=0, b=1, lty=2)                                                       # droite de certitude
abline(a=crossprod(X[2,], X[1,])/X[2,2], b=-X[2,1]/X[2,2])                    # loteries équitables
lines(c[97500 > 3*c^2], sqrt(97500-3*c[97500 > 3*c^2]^2), col="red", lwd=2)   # courbe d'indifférence
```

On constate que dans ce cas, la courbe d'indifférence est concave. En effet, **quand l'agent aime le risque, sa fonction d'utilité est convexe et ses courbes d'indifférence sont concaves**.

Enfin, pourquoi est-ce que se déplacer sur la droite des loteries équitables est équivalent à jouer à une loterie d'espérance nulle ? Notons que toute loterie $X'$ de même espérance que $X$ (c'est-à-dire n'importe quelle loterie sur la droite des loteries équitables passant par $X$) peut s'écrire

\begin{equation}
    X' = X + (X' - X)
\end{equation}

avec $\mathbf E(X' - X) = \mathbf E(X) - \mathbf E(X') = 0$, c'est-à-dire que la loterie $X' - X$ est d'espérance nulle. Pour un agent qui possède initialement la loterie $X$, choisir $X'$ est donc équivalent à jouer à la loterie d'espérance nulle $X' - X$.
