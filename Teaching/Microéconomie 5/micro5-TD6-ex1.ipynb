# Microéconomie 5
## TD6 : Risque et incertain

### Exercice 1

On étudie le comportement de deux agents, Annalise ($A$) et Bonnie ($B$), face au risque.

On commence avec $B$. Trois points de sa fonction d'utilité sont connus,

\begin{aligned}
    v^B(0) = 0
    \qquad
    v^B(85\,000) = 0{,}7
    \qquad
    v^B(100\,000) = 1
\end{aligned}

et on souhaite savoir si elle préfère conserver sa richesse initiale, 85 000 €, ou miser cette richesse pour gagner 15 000 € de plus avec probabilité 0,8 ou tout perdre avec probabilité 0,2. Formellement, $B$ doit donc choisir entre les deux loteries $L_0$ et $L_1$ suivantes :

\begin{equation}
    L_0 = (85\,000, 1)
    \qquad
    L_1 = (100\,000, 0{,}8 ; 0, 0{,}2).
\end{equation}

Pour connaître son choix, on calcule l'utilité espérée qu'elle tire de ces deux lotteries :

\begin{aligned}
    U^B(L_0) &= 1 \times \overbrace{v^B(85\,000)}^{= 0{,}7} = 0{,}7 \\
    U^B(L_1) &= 0{,}8 \times \underbrace{v^B(100\,000)}_{= 1} + 0{,}2 \times \underbrace{v^B(0)}_{= 0} = 0{,}8.
\end{aligned}

$U^B(L_1) > U^B(L_0)$ donc $B$ choisit la loterie risquée $L_1$.

Que peut-on dire sur l'attitude de $B$ face au risque ? Puisque $\mathbf E(L_1) = 80\,000 < 85\,000 = \mathbf E(L_0)$, on peut affirmer que $B$ aime le risque : en effet, elle préfère une loterie risquée ($L_1$) à une loterie certaine d'espérance supérieure ($L_0$). Formellement,

\begin{equation}
    U^B(L_1) > U^B(L_0) = v^B(85\,000) > v^B(80\,000) = U^B(\mathbf E(L_1))
\end{equation}

où les loteries $L_1$ et $\mathbf E(L_1)$ ont la même espérance, mais $L_1$ est risquée et $\mathbf E(L_1)$ est certaine. \[Notez que, par abus de notation, j'appelle $\mathbf E(L_1)$ la loterie certaine qui donne $\mathbf E(L_1)$ à tous les coups, $(\mathbf E(L_1), 1)$.\]

Représentons cette situation sur un graphe.

```r
c <- seq(0, 110000, length=1000)
v0 <- 0
v85000 <- 0.7
v100000 <- 1

options(scipen=999)            # enlever notation scientifique, e.g. 2E4 pour 20000
plot(x = c, y = c, xlim = c(0, 110000), ylim = c(0, 1.2),
     type = "n", xlab = "Richesse", ylab = "Utilité")
points(c(0, 85000, 100000), c(v0, v85000, v100000), pch=19, col="red")
lines(c, (c/100000)^2.2, col="red", lty=2, lwd=1)
lines(c(0, 100000), c(0, 1), lty=2, lwd=1)
points(80000, .8, col="red")

points(c(0, 85000, 100000), c(v0, v85000, v100000), pch=19, col="red")
```

Ci-dessus, les trois points rouges pleins sont les points connus de la fonction d'utilité de $B$, $v^B$. En pointillés, j'ai représenté un exemple de fonction d'utilité qui pourrait convenir (i.e. qui passe par ces trois points).

En pointillés noirs, j'ai tracé le segment qui permet de trouver graphiquement $U^B(L_1)$. En effet, l'utilitée procurée par $L_1$ est une somme pondérée (par les probabilités respectives) de $v^B(100\,000)$ et $v^B(0)$, donc se situe sur le segment qui relie ces deux points (plus exactement, aux 4/5èmes de ce segment puisqu'il y a 4 chances sur 5 de gagner 100 000 €).

Le point rouge vide représente donc l'utilité $U^B(L_1)$. On voit clairement que ce point est plus haut que le point rouge qui dénote $v^B(85\,000)$, ce qui signifie que $B$ préfère $L_1$ à $L_0$ et, par suite, permet d'affirmer que $B$ aime le risque.

Étudions maintenant le cas de $A$, qui assure qu'à la place de $B$ elle aurait refusé la loterie $L_1$. Plus encore, elle affirme qu'elle n'aurait accepté cette loterie qu'à partir du moment où les pertes se seraient limitées à 35 000 € en cas d'échec – c'est-à-dire que sa richesse finale en cas d'échec serait 50 000 €.

Puisque la fonction d'utilité $v^A$ de $A$ est croissante, cela veut dire qu'elle aurait refusé toute loterie qui lui donne moins de 50 000 € en cas d'échec, et accepté toute loterie qui lui donne plus de 50 000 € en cas d'échec. Par continuité de la fonction $v^A$, $A$ est donc indifférente entre la loterie certaine $L_0$ et la loterie $L_2$ suivante :
\begin{equation}
    L_2 = (100\,000, 0{,}8 ; 50\,000, 0{,}2).
\end{equation}

Que peut-on en déduire sur l'attitude face au risque de $A$ ? Il faut déterminer si $A$ préfère la loterie risquée $L_2$ ou la loterie équitable mais certaine $\mathbf E(L_2)$. Puisque $\mathbf E(L_2) = 90\,000$,

\begin{equation}
    U^A(L_2) = U_A(L_0) = v^A(85\,000) < v^A(90\,000) = U^A(\mathbf E(L_2)).
\end{equation}

Ainsi, **entre les deux loteries équitables $L_2$ et $\mathbf E(L_2)$, $A$ préfère la loterie certaine $\mathbf E(L_2)$ à la loterie risquée $L_2$** : $A$ est donc averse au risque.
