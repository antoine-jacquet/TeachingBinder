### Exercice 5 : Partage optimal du risque

Il y a deux états de la nature $\{1, 2\}$ qui se réalisent avec probabilité $\pi_1$ et $\pi_2$ respectivement ($\pi_1 + \pi_2 = 1$). Il y a un risque macroéconomique : les ressources disponibes sont $\omega_1 = 100$ dans l'état 1, et $\omega_2 = 200$ dans l'état 2.

On cherche comment allouer de façon Pareto-optimale les biens contingents 1 et 2 à deux agents $A$ et $B$, dont les préférences sont représentées par les fonctions d'utilité $v^A$ et $v^B$. En particulier, on se demande si les deux allocations suivantes sont optimales :

1. $c^A = (50, 50)$ et $c^B = (100, 100)$,
2. $c^A = (10, 120)$ et $c^B = (90, 80)$.

```r
omega1 <- 100
omega2 <- 200
```

**Premier cas.** On suppose d'abord que les fonctions d'utilité sont définies sur $\mathbf R_+$ par $v^A(c) = v^B(c) = \ln c$.

Les optima de Pareto internes doivent donc vérifier

\begin{equation}
    \mathsf{TMS}^A = \mathsf{TMS}^B
    \iff
    \underbrace{\frac{\pi_1 (v^A)'(c_1^A)}{\pi_2 (v^A)'(c_2^A)} = \frac{\pi_1 (v^B)'(c_1^B)}{\pi_2 (v^B)'(c_2^B)}}_\text{condition de Borch}
    \iff
    \boxed{ \frac{c_2^A}{c_1^A} = \frac{c_2^B}{c_1^B} }.
\end{equation}

On vérifie facilement que l'allocation 1 satisfait la condition d'optimalité ci-dessus : $\frac{50}{50} = 1 = \frac{100}{100}$. Mais cette condition n'est _pas suffisante_ : pour être optimale, il faut aussi qu'une allocation respecte les conditions de ressources $c_1^A + c_1^B = \omega_1$ et $c_2^A + c_2^B = \omega_2$. Or ce n'est pas le cas pour l'allocation 1 : $50 + 100 > 100$ et $50 + 100 < 200$ : il y a plus de bien 1 alloué qu'il n'y en a de disponible, et au contraire tout le bien 2 disponible n'est pas alloué. L'allocation 1 n'est donc _pas optimale_.

L'allocation 2 ne satisfait quant à elle pas la condition d'optimalité trouvée ci-dessus : $\frac{120}{10} \neq \frac{80}{90}$, elle n'est donc _pas optimale_ non plus.

Trouvons maintenant l'expression de la courbe des optima internes. On part de la condition de Borch puis on utilise les conditions de ressources $c_1^A + c_1^B = \omega_1$ et $c_2^A + c_2^B = \omega_2$ pour trouver une expression de $c_2^A$ en fonction de $c_1^A$ :

\begin{equation}
    \frac{c_2^A}{c_1^A} = \frac{c_2^B}{c_1^B} = \frac{\omega_2 - c_2^A}{\omega_1 - c_1^A}
    \iff
    \boxed{ c_2^A = \frac{\omega_2}{\omega_1} c_1^A } \overset{\text{A.N.}}{=} 2 c_1^A.
\end{equation}

Un exemple d'allocation optimale aurait donc été $c^A = (25, 50)$ et $c^B = (75, 150)$.

Je trace maintenant cette courbe des optima internes (en violet).

```r
c1 <- seq(0, omega1, length=1000)
c2_optim <- omega2/omega1*c1

plot(x = c1, y = c1, xlim = c(0, omega1), ylim = c(0, omega2),
     type = "n", xlab = "Bien contingent 1", ylab = "Bien contingent 2", xaxs = "i", yaxs = "i")
lines(c1, c2_optim, col="purple", lwd=2)
```

**Deuxième cas.** On suppose désormais que les fonctions d'utilité sont définies sur $\mathbf R_+$ par $v^A(c) = \ln c$ et $v^B(c) = 10 + c$.

Les optima de Pareto internes doivent maintenant vérifier

\begin{equation}
    \mathsf{TMS}^A = \mathsf{TMS}^B
    \iff
    \underbrace{\frac{\pi_1 (v^A)'(c_1^A)}{\pi_2 (v^A)'(c_2^A)} = \frac{\pi_1 (v^B)'(c_1^B)}{\pi_2 (v^B)'(c_2^B)}}_\text{condition de Borch}
    \iff
    \frac{c_2^A}{c_1^A} = \frac{1}{1}
    \iff
    \boxed{ c_2^A = c_1^A }.
\end{equation}

Encore une fois, l'allocation 1 satisfait la condition d'optimalité mais pas les conditions de ressources, donc elle n'est pas optimale. L'allocation 2 ne satisfait pas la condition d'optimailité donc n'est pas optimale non plus.

Cette fois, la condition de Borch nous a immédiatement donné $c_2^A$ en fonction de $c_1^A$, c'est-à-dire l'équation de la courbe des optima de Pareto internes : $c_2^A = c_1^A$.

Un exemple d'allocation optimale aurait donc été $c^A = (50, 50)$ et $c^B = (50, 150)$.

Je trace maintenant la courbe des optima internes (en violet, trait plein), ainsi que les optima en coin à titre informatif (en pointillés).

```r
c1 <- seq(0, omega1, length=1000)
c2_optim <- c1

plot(x = c1, y = c1, xlim = c(0, omega1), ylim = c(0, omega2),
     type = "n", xlab = "Bien contingent 1", ylab = "Bien contingent 2", xaxs = "i", yaxs = "i")
lines(c1, c2_optim, col="purple", lwd=2)
lines(c(omega1, omega1)-.005*omega1, c(max(c2_optim), omega2), col="purple", lty=2, lwd=2)                               # courbe des optima internes
```
